```python
count = 0
for i in df["no_of_employees"]:
  if i < 0:
    count = count + 1
count
```

If the numbers of employees less than zero means their is no meaning to it. It should be replaced with zero.

```python
df.loc[df["no_of_employees"] < 0, "no_of_employees"] = 0
```
